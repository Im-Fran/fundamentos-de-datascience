Calcula el promedio de notas de cada estudiante y determina quien tiene el promedio más alto y más bajo

```python
import pandas as pd

# Crear el DataFrame para modelar con pandas
df_estudiantes = pd.DataFrame(estudiantes)

# Calcular el promedio de notas para cada estudiante
promedio = lambda notas: sum(notas) / len(notas)
df_estudiantes['promedio'] = df_estudiantes['notas'].apply(promedio)

# Encontrar el promedio más alto y más bajo
promedio_mas_alto = df_estudiantes['promedio'].max()
promedio_mas_bajo = df_estudiantes['promedio'].min()

print(f'Promedio mínimo de las notas es: {promedio_mas_bajo:.2f}')
print(f'Promedio máximo de las notas es: {promedio_mas_alto:.2f}')
```
