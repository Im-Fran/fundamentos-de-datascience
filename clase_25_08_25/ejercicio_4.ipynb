Tienes una lista de estudiantes, cada uno representado por un diccionario con su nombre y lista de notas.

```python
# datos
estudiantes = [
    { 'nombre': 'Ana', 'notas': [6.5, 7.0, 5.8] },
    { 'nombre': 'Luis', 'notas': [4.2, 5.1, 6.0] },
    { 'nombre': 'Sofía', 'notas': [3.9, 4.0, 4.5] },
    { 'nombre': 'Carlos', 'notas': [5.2, 6.1, 4.8] },
    { 'nombre': 'María', 'notas': [6.7, 5.9, 6.3] },
    { 'nombre': 'Javier', 'notas': [4.5, 4.8, 5.0] },
    { 'nombre': 'Lucía', 'notas': [5.9, 6.2, 6.0] },
    { 'nombre': 'Pedro', 'notas': [3.8, 4.1, 4.0] },
    { 'nombre': 'Elena', 'notas': [6.0, 5.7, 6.5] },
    { 'nombre': 'Raúl', 'notas': [4.3, 4.9, 5.2] },
    { 'nombre': 'Valeria', 'notas': [5.5, 6.0, 5.8] },
    { 'nombre': 'Miguel', 'notas': [4.0, 4.2, 4.5] },
    { 'nombre': 'Isabel', 'notas': [6.1, 6.3, 6.7] },
    { 'nombre': 'Diego', 'notas': [5.0, 5.2, 5.5] },
    { 'nombre': 'Paula', 'notas': [4.7, 5.1, 5.3] },
    { 'nombre': 'Andrés', 'notas': [3.9, 4.0, 4.2] },
    { 'nombre': 'Sara', 'notas': [6.4, 6.6, 6.8] },
    { 'nombre': 'Fernando', 'notas': [5.3, 5.7, 5.9] },
    { 'nombre': 'Natalia', 'notas': [4.6, 4.8, 5.0] },
    { 'nombre': 'Alberto', 'notas': [5.8, 6.0, 6.2] },
    { 'nombre': 'Rocío', 'notas': [4.1, 4.3, 4.5] },
    { 'nombre': 'Iván', 'notas': [6.2, 6.4, 6.6] },
    { 'nombre': 'Patricia', 'notas': [5.4, 5.6, 5.8] },
    { 'nombre': 'Hugo', 'notas': [4.9, 5.1, 5.3] },
    { 'nombre': 'Carmen', 'notas': [6.0, 6.2, 6.4] },
    { 'nombre': 'Mario', 'notas': [4.4, 4.6, 4.8] },
    { 'nombre': 'Julia', 'notas': [5.7, 5.9, 6.1] },
    { 'nombre': 'Adrián', 'notas': [4.2, 4.4, 4.6] },
    { 'nombre': 'Marta', 'notas': [6.3, 6.5, 6.7] },
    { 'nombre': 'Esteban', 'notas': [5.1, 5.3, 5.5] },
    { 'nombre': 'Clara', 'notas': [4.8, 5.0, 5.2] },
    { 'nombre': 'Gabriel', 'notas': [6.5, 6.7, 6.9] }
]
```

# Desafío
1. Calcula el promedio de notas de cada estudiante y determina quien tiene el promedio más alto y más bajo
2. Cuenta cuántos estudiantes aprobaron todas sus asignaturas (todas las notas >= 4.0)
3. ¿Cuál es la nota más frecuente (moda) considerando todas las notas de todos los estudiantes?
4. ¿Qué porcentaje de estudiantes tiene al menos una nota bajo 4.0?
5. Entrega un listado ordenado (de mayor a menor) de los estudiantes según su promedio.

```python
# Datos
est_min = {'promedio': None} # Estudiante con promedio más bajo
est_max = {'promedio': None} # Estudiante con promedio más alto
cantidad_aprobados = 0 # Cantidad de estudiantes que aprobaron todas sus asignaturas
cantidad_nota_al_menos_una_bajo_4 = 0 # Cantidad de estudiantes con al menos una nota bajo 4.0
frecuencias = {}

# Insertar en cada estudiante su promedio
for estudiante in estudiantes:
    notas = estudiante['notas']
    promedio = round(sum(notas) / len(notas), 2)
    estudiante['promedio'] = promedio

    for nota in notas:
        frecuencias[nota] = frecuencias.get(nota, 0) + 1

    if min(notas) >= 4.0: # Si existe alguna nota menor a 4.0, no aprueba y se salta
        cantidad_aprobados += 1
    else:
        cantidad_nota_al_menos_una_bajo_4 += 1

    if est_max['promedio'] is None:
        est_max = estudiante
    elif est_max['promedio'] <= promedio:
        est_max = estudiante

    if est_min['promedio'] is None:
        est_min = estudiante
    elif est_min['promedio'] >= promedio:
        est_min = estudiante


print(f"Estudiante con promedio más alto: {est_max['nombre']} con promedio {est_max['promedio']}")
print(f"Estudiante con promedio más bajo: {est_min['nombre']} con promedio {est_min['promedio']}")
print(f"Cantidad de estudiantes que aprobaron todas sus asignaturas: {cantidad_aprobados}")

nota_mas_frecuente = max(frecuencias, key=frecuencias.get)
cantidad_de_veces_nota_mas_frecuente = frecuencias[nota_mas_frecuente]
print(f"La nota más frecuente (moda) es: {nota_mas_frecuente}, con una frecuencia de {cantidad_de_veces_nota_mas_frecuente}")

porcentaje_estudiantes_con_al_menos_una_nota_bajo_4 = round((cantidad_nota_al_menos_una_bajo_4 / len(estudiantes)) * 100, 2)
print(f"Porcentaje de estudiantes con al menos una nota bajo 4.0: {porcentaje_estudiantes_con_al_menos_una_nota_bajo_4}%")

# Ordenar estudiantes por promedio de mayor a menor
# lambda est: est['promedio'] es una función anónima que toma un estudiante y devuelve su promedio
# reverse=True para ordenar de mayor a menor (porque por defecto es de menor a mayor)
estudiantes_ordenados = sorted(estudiantes, key=lambda est: est['promedio'], reverse=True)
print("Listado ordenado de estudiantes según su promedio (de mayor a menor):")
for estudiante in estudiantes_ordenados:
    print(f"{estudiante['nombre']}: {estudiante['promedio']}")
```
