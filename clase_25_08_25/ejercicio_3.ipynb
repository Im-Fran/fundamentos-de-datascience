# Ejercicio 3
Dado un listado de notas, calcula cuántos estudiantes están en cada tramo: reprobados (<4.0), aprobados (4.0-5.9) y destacados (>= 6.0). Además, muestra el promedio de cada grupo y el porcentaje que representan del total de estudiantes.

```python
# Datos
notas = [4.8, 6.2, 5.5, 3.9, 7.0, 4.1, 5.8, 6.0, 3.5, 5.2, 6.8, 2.9, 4.0, 5.0, 6.5]

def calc():
    # Reprobados
    reprobados = [nota for nota in notas if nota < 4.0]
    largo_reprobados = len(reprobados)
    promedio_reprobados = (sum(reprobados) / largo_reprobados)
    porcentaje_reprobados = (largo_reprobados / len(notas)) * 100

    # Aprobados
    aprobados = [nota for nota in notas if 4.0 <= nota <= 5.9]
    largo_aprobados = len(aprobados)
    promedio_aprobados = (sum(aprobados) / largo_aprobados)
    porcentaje_aprobados = (largo_aprobados / len(notas)) * 100

    # Destacados
    destacados = [nota for nota in notas if nota >= 6.0]
    largo_destacados = len(destacados)
    promedio_destacados = (sum(destacados) / largo_destacados)
    porcentaje_destacados = (largo_destacados / len(notas)) * 100


    print(f"Reprobados: {largo_reprobados}")
    print(f"Promedio: {promedio_reprobados:.2f}")
    print(f"Porcentaje: {porcentaje_reprobados:.2f}%")
    print("-----------------------")

    print(f"Aprobados: {largo_aprobados}")
    print(f"Promedio: {promedio_aprobados:.2f}")
    print(f"Porcentaje: {porcentaje_aprobados:.2f}%")
    print("-----------------------")

    print(f"Destacados: {largo_destacados}")
    print(f"Promedio: {promedio_destacados:.2f}")
    print(f"Porcentaje: {porcentaje_destacados:.2f}%")


# Validar lista tiene datos y son números (revisando el tipo)
if len(notas) > 0:
    calc()
else:
    print("La lista de notas está vacía o contiene datos no numéricos.")
```
