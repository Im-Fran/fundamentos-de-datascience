# Trabajando con Matplotlib

```python
import pandas as pd
import matplotlib as mp

csv = pd.read_csv("../datos_de_prueba/Estudiantes.csv")

notas = ['nota_1', 'nota_2', 'nota_3']
df_estudiantes = pd.DataFrame(csv)
df_estudiantes['promedio'] = df_estudiantes[notas].mean(axis=1)
df_estudiantes.head()

# Muestra los datos con Matplotlib
df_estudiantes.plot(x='nombre', y='promedio', kind='bar', title='Promedio de Notas por Estudiante', legend=False)

# Muestra la moda de las notas en matplotlib
df_estudiantes[notas].mode().plot(kind='bar', title='Moda de Notas', legend=True)
```
