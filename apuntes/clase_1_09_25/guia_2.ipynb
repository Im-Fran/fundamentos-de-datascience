# Ejercicio 3: Indexación y Slicing

Crea una matriz 3x3 con números del 1 al 9. A partir de ella, extrae:

a) El número en la segunda fila, tercera columna (el 6).

b) La segunda fila completa.

c) La tercera columna completa.

```python
print(np.arange(1,10))
matriz = np.arange(1, 10).reshape(3, 3)
print("Matriz:\n", matriz)

# a) Número en la segunda fila, tercera columna
num_a = matriz[1, 2]
print("Número en la segunda fila, tercera columna:", num_a)

# b) Segunda fila completa
fila_b = matriz[1, :]
print("Segunda fila completa:", fila_b)

# c) Tercera columna completa
columna_c = matriz[:, 2]
print("Tercera columna completa:", columna_c)
```

# Ejercicio 4: Estadísticas Básicas
A partir del array de notas `notas = np.array([4.5, 6.2, 3.9, 7.0, 5.5, 2.1])`, calcula e imprime el promedio, la desviación estándar, la nota máxima, y la nota mínima.

```python
import numpy as np

notas = np.array([4.5, 6.2, 3.9, 7.0, 5.5, 2.1])

promedio = notas.mean()
desviacion_estandar = notas.std()
nota_maxima = notas.max()
nota_minima = notas.min()

print(f"Promedio: {promedio:.2f}")
print(f"Desviación Estándar: {desviacion_estandar:.2f}")
print("Nota Máxima:", nota_maxima)
print("Nota Mínima:", nota_minima)
```

Ejercicio 6: Inyectar notas aprobadas (notas >= 4.0)

```python
notas = np.array([4.5, 6.2, 3.9, 7.0, 5.5, 2.1])
notas_aprobadas = notas[notas >= 4.0]

print("Notas Aprobadas:", notas_aprobadas)
```
