# Ejercicio 2
Calcular el promedio, la nota más alta, la más baja y cuántos estudiantes obtuvieron sobre 5.0

```python
notas = [4.8, 6.2, 5.5, 3.9, 7.0, 4.1, 5.8, 6.0, 3.5, 5.2]

# Sobresalientes
contador_sobresalientes = 0
for nota in notas:
    if nota > 5.0:
        contador_sobresalientes += 1

# Algoritmo
print(f"La nota más alta es {max(notas)}")
print(f"La nota más baja es {min(notas)}")
print(f"El promedio del curso es {sum(notas) / len(notas):.2f}")
print(f"Cantidad de estudiantes con nota sobre 5.0: {contador_sobresalientes}")
```
