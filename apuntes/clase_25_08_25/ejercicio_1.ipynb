Problema: Calcula el promedio de un curso y cuenta cuantos aprobaron

```python
# Datos
notas_curso = [6.5, 7.0, 3.2, 4.9, 5.8, 2.1, 4.0]
nota_aprobacion = 4.0
suma_notas_aprobadas = 0
suma_notas = 0
```

Filtro de notas aprobadas

```python
suma_notas = 0
suma_notas_aprobadas = 0
for nota in notas_curso:
    suma_notas += nota
    if nota >= nota_aprobacion:
        suma_notas_aprobadas +=1
        print(f"Nota aprobada: {nota}")


print(f"Total de notas aprobadas: {suma_notas_aprobadas}")
print(f"Promedio del curso: {(suma_notas / len(notas_curso)):.2f}")
```
